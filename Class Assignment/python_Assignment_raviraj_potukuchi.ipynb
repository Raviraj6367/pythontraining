## 21. Flatten nested list (recursion)

```python
#Solution already provided
# go through the code and see if you understand it,
# Write the logic used for each line
def flatten(lst):
    res=[] # empty list created
    for x in lst: # traversing through the 3D list
        if isinstance(x,list): res.extend(flatten(x))
        # # For each item in 3D list if the item is a list then recursive call is made

        else: res.append(x)  # else if only single item then directly appended to the list
    return res # after traversing all the elements in the 3D list the final res list is returned
print(flatten([1,[2,3],[4,[5,6]]]))

# 1--> 1
# 2 --> [2,3]-->res = [1,2,3,4,5,6]
#3--> [4,[5,6]]


```

```python
#Solution already provided
# go through the code and see if you understand# Write the logic used for each line. Solution and explanation both provided, Go through it to see

def subsets(lst):
    if not lst: return [[]] #checks list is empty or not.
    x=lst[0] #x=1
    rec=subsets(lst[1:]) # Recursively computes all subsets of the remaining lists.
    return rec + [[x]+r for r in rec]
    # rec already has all the subsets that do not include x.
    #The list [[x]+r for r in rec] --> loops over every subset r in rec.  --> Creates a nuw subset by putting x in front --> [x]+r
    # So for rec =[[], [2]] and x=1;
    # for r = [],--> [x]+r = [1]+[] = [1]
    # for r =[2]-> [x]+ r= [1]+[2]= [1,2]
    # So the final list comprehension becomes [[1], [1,2]]

    # rec + [[x] +r for r in rec]:
    # Concatenates subsets without x(rec) and subsets with x (new ones).
    # [[], [2],[1],[1,2]]
print(subsets([1,2,3]))

#Logic
#This code defines a recursive function subsets that generates all unique subsets (also known as the power set) of a given list. Let's break down the logic:

#def subsets(lst): defines the function subsets that takes one argument, lst, which is the input list.

#if not lst: return [[]]: This is the base case of the recursion. If the input list lst is empty (meaning not lst is True), it returns a list containing a single empty list [[]]. This represents the only subset of an empty set, which is the empty set itself.

#x=lst[0]: If the list is not empty, this line takes the first element of the list and assigns it to the variable x.

#rec=subsets(lst[1:]): This is the recursive step. It calls the subsets function again, but this time with a new list lst[1:], which is the original list without its first element (x). The result of this recursive call (all subsets of the rest of the list) is stored in the variable rec.

#return rec + [[x]+r for r in rec]: This line combines the results to form the final set of subsets. It does two things:

#It takes all the subsets found in rec (subsets that don't include x).
#It then creates a new set of subsets by taking each subset r from rec and adding x to the beginning of it (represented by [x]+r). These are the subsets that do include x. By concatenating these two sets of subsets (rec and the new list comprehension), it returns all possible unique subsets of the original list.
#Finally, print(subsets([1,2])) calls the function with the list [1,2]. Let's trace it:

#subsets([1,2])
#x = 1
#rec = subsets([2])
#x = 2
#rec = subsets([])
#Base Case: returns [[]]
#Returns [[]] + [[2]+[] for [] in [[]]] which is [[]] + [[2]] -> [[], [2]]
#Returns [[], [2]] + [[1]+r for r in [[], [2]]] which is [[], [2]] + [[1], [1, 2]] -> [[], [2], [1], [1, 2]]
#The output will be [[], [2], [1], [1, 2]] (the order might vary slightly depending on concatenation, but the elements will be the same).
```

## 23. Fibonacci memoization

```python
# go through the code and see if you understand# Write the logic used for each line
#Calculate the nth Fibonacci number using recursion
memo={}
def fib(n):
    if n<=1: return n
    if n in memo: return memo[n]
    memo[n]=fib(n-1)+fib(n-2)
    return memo[n]
print(fib(10))
```

## 24. Custom sorting by

```python
# Explain the code like this for the below ones.
#This code sorts a list of words. First, it initializes a list words with five strings.
# Then, words.sort(key=lambda w:(len(w),w)) sorts this list in place.
#The key argument uses a lambda function to define how each word should be compared. For every word w, it generates a tuple (len(w), w). Python sorts tuples by comparing their elements from left to right. So, it first sorts the words by their length (shortest first). If two words have the same length, it then sorts them alphabetically based on the word itself. Finally, print(words) displays the sorted list.
words=["apple","car","banana","kiwi","ball"]
words.sort(key=lambda w:(len(w),w))
print(words)
```

#32. Create Word Frequency Dictionary


```python
#Use: reduce() + lambda
sentence ="nlp is fun and nlp is powerful"
#Ex: {'nlp': 2}
sentence= sentence.split(' ')
from collections import Counter
dic = {k:v for k, v in Counter(sentence).most_common(len(sentence))}
print(dic)
# Reduce and Lambda

freq = reduce(lambda acc, word : {**acc, word : acc.get(word,0) +1 }, sentence, {})
print(freq)
```
