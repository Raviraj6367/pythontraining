# **Demo 01: Working with Lists in Python**

**Objective:** To create, modify, and access elements in Python lists using indexing, slicing, and built-in methods

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Create a list containing at least 10 elements of different data types (integers, strings, and floats)
2. Access and print elements using positive indexing from the list
3. Access and print elements using negative indexing from the list
4. Perform slicing operations to extract a sublist from the original list
5. Append a new element and remove an existing element from the list


**Step 1:** Create a list containing at least 10 elements of different data types (integers, strings, and floats)

```python
my_list = [25, "Hello", 3.14, "Python", 100, 7.89, "AI", 42, "Data", 2.718, True] # list starts and ends with []
my_list
```

```python
my_list[0]
```

```python
my_list[4]
```

```python
my_list[-1]
```

**Step 2:** Access and print elements using positive indexing from the list by printing 1st, 5th and Last Element of the list.

```python
positive_indexing = [my_list[0], my_list[4], my_list[-1]]
print(positive_indexing)
```

**Step 3:** Access and print elements using negative indexing from the list

```python
negative_indexing = [my_list[-1], my_list[-4], my_list[-7]]
print(negative_indexing)
```

**Step 4:** Perform slicing operations to extract a sublist from the original list

```python
sublist = my_list[2:7] # syntax --> list[ starting_index : ending_index+1]. # 0,5. --> 0,1,2,3,4
print(sublist) # print all values in range 2nd index till 6th index
```

```python
print(my_list)
print(my_list[0:])
print(my_list[0:11]) # 0 to 10
```

```python
#Print every 3nd element in the list
print(my_list[0:11:3] )
print(my_list[::3]) # Start from 1st go till last and print every 3rd element
print(my_list[::])##Start from 1st go till last and print every element
```

```python
#Start from the 2nd element and print every 2nd element till the 2nd last element,.


print(my_list[1:-1:1])
```

```python
print(my_list[-2:0:-1])
```

```python
print(my_list)

#print(my_list[-1:-12:-1]) # list[starting_indx :  ending_index+1 : step_size ] +ve step size means --> left to right.
#-ve step size means, setp is from right to left.
```

```python
my_list[::-1]
```

```python
print(my_list[3:-1:2])
```

```python
print(my_list[-1:-12:-3])
```

```python
# Print the list --> 2nd Last Element till 2nd element in reverse order.
print(my_list[-2:0:-1])
```

```python
# Print the list --> 2nd Last Element till 1st element in reverse order.
print(my_list[-2::-1])
```

```python
#Print the list from 3rd Element till 3rd Last Element --> Original Order and Reverse Order.
print(my_list)
print(my_list[2:-2])
print(my_list[-3:1:-1])
```

**Step 5:** Append a new element and remove an existing element from the list

```python
my_list = [25, "Hello", 3.14, "Python", 100, 7.89, "AI", 42, "Data", 2.718]
print(my_list)
my_list.append("New Element")
print(my_list)

```

```python
my_list_2= ['1,2,3']
my_list.append(my_list_2)
print(my_list)
```

```python
my_list_2= ['1,2,3']
my_list.extend(my_list_2)
print(my_list)
```

```python
type(my_list[-2])
```

```python
type(my_list[-1])
```

```python
my_list.remove('1,2,3') # Directly remove the element from list using remove function.
print(my_list)
```

```python
my_list[-1] = '1234'
print(my_list)
```

```python
my_list.count("Python")
```
