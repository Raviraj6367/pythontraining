```python
# 1. Creating a dictionary with at least 5 key-value pairs (student details)
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

print(student_dict)
```

**Step 2:** Access and print values using keys

```python
student_name = student_dict["name"] # map.get(key_name)
student_course = student_dict["course"]

print(student_name)
print(student_course)
```

**Step 3:** Modify an existing value in the dictionary

```python
student_dict["marks"] = 90  # Updating marks
print(student_dict)
```

**Step 4:** Add a new key-value pair and remove an existing key

```python
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

student_dict["email"] = "johndoe@example.com"  # Adding email
student_dict
```

```python
print(student_dict.pop("city") ) # Restudent_dictmoving city and also want to see the key-value pair being removed.

```

```python
student_dict
```

```python
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}
print(student_dict.pop("city") ) # Removing city
print(student_dict)
```

**Step 5:** Retrieve all keys, values, and key-value pairs using dictionary methods

```python
all_keys = list(student_dict.keys()) # Extract all the keys of the dictionary
all_values = list(student_dict.values()) # All the values from the dictionary and storing as a list
all_items = list(student_dict.items()) # Extract all the items from the dict and store it as a list

print(all_keys)
print(all_values)
print(all_items)
```

**Step 6:** Use the get() method to access a key that may or may not exist

```python
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

existing_value = student_dict.get("course1")  # Returns "Computer Science"
existing_value
```

```python
non_existing_value = student_dict.get("phone", "Key not found")  # Returns "Key not found"
non_existing_value

```

```python
# Use zip() to create dictionaryes
keys = ['a', 'b', 'c',1,2,3,4,45,5,6,7,8,9,9]
values = (1, 2, 3,1,2,3,4,4,5,5,6,6,9,90)
var1 = zip(keys,values)
for element in var1:
  print(element)
```

```python
col_list = ['Name','Age','Marks', 'Class']
val_list= [['Amit','Ajit','Nikunj'], (12,13,14), [300,400,500], [1,2,3]]

name_dict = dict(zip(col_list, val_list)) # When to use: in case of parallel data, ETL/preprocessing of Data.
print(name_dict)
```
