**Objective:** To understand how sets work, including adding, removing, and performing mathematical set operations

**Prerequisites:** None

**Tools required:** Python

#### **Steps to be followed:**

1. Create a set with at least seven unique elements, including numbers and strings
2. Add a new element to the set and remove an existing one
3. Try adding a duplicate element and observe the result
4. Perform union, intersection, and difference operations with another set
5. Use a set to remove duplicates from a list
6. Convert a set into a list and a tuple


```python
my_set = {10:10}
type(my_set)
```

**Step 1:** Create a set with at least seven unique elements, including numbers and strings

```python
my_set = {10, "Python", 3.14, "AI", 42, "Data", 7.89, 10} # Mutable Unordered and unique and unindexed
print(my_set)
```

```python
#zip --> club two data structures together to form a dictionary.,
```

**Step 2:** Add a new element to the set and remove an existing one

```python
my_set.add("Machine Learning")  # Adding a new element
print(my_set)
my_set.remove(42)  # Removing an existing element
print(my_set)
```

**Step 3:** Try adding a duplicate element and observe the result

```python
my_set.add("Python")  # "Python" already exists, so no change
print(my_set)
```

**Step 4:** Perform union, intersection, and difference operations with another set

```python
another_set = {3.14, "AI", "Deep Learning", 99, "Data"}
print(my_set)
print(another_set)
union_set = my_set.union(another_set)  # Combines unique elements from both sets
intersection_set = my_set.intersection(another_set)  # Common elements
difference_set = my_set.difference(another_set)  # Elements in my_set but not in another_set
symm_diff = my_set.symmetric_difference(another_set) #  Gives the elements that are present in either set but NOT in both.
print(union_set)
print(intersection_set)
print(difference_set)
print(symm_diff)
```

**Step 5:** Use a set to remove duplicates from a list

```python
duplicate_list = [10, "Python", 3.14, "AI", 10, "Python", "AI", 42]
print(duplicate_list)
unique_list = list(set(duplicate_list))  # Converts the list to a set and back to a list to remove duplicates
print(unique_list)
```

```python
list1 = [1,2,3,4,5,5,5]
list(set(list1))
```

**Step 6:** Convert a set into a list and a tuple

```python
set_as_list = list(my_set)  # Set converted to a list (unordered)
set_as_tuple = tuple(my_set)  # Set converted to a tuple (unordered but immutable)
print(set_as_list)
print(set_as_tuple)
```
