# __Arbitrary Arguments__

### __Non-keyword arbitrary argument(*args)__

- The function total is designed to accept a variable number of non-keyword arguments using __*args__.
- The __*args__ parameter allows the function to take any number of arguments, which are passed as a tuple.
- You can then iterate over them using a __for__ loop.

```python
# Define the function 'total' which takes a variable number of non-keyword arguments
def total(*aahdfjhadjkhad):
    # Initialize a variable 'tot' to store the total
    tot = 0
    # Iterate over the arguments and add each to 'tot'
    for n in aahdfjhadjkhad:
        tot += n
    # Print the total
    print(tot)

# Call the 'total' function with two arguments: 23 and 67
total(23, 67)
```

- This function call passes five numbers to the function, and the function sums them.

```python
# Call the 'total' function with five arguments: 90, 100, 10, 30, 40
total(90, 100, 10, 30, 40)
```

```python

def list_adder(*args):
  total=0
  for lists in args:
    print('The list is', lists)
    for item in lists:
      print('The Item is ', item)
      total+=item
  return total
```

### __Keyword arbitrary argument__

- The function information is designed to accept a variable number of keyword arguments using __**kwargs__.
- The __**kwargs__ parameter allows the function to take any number of keyword arguments.
- These arguments are passed as a dictionary, where the keys are the parameter names and the values are the corresponding argument values.
- The function uses a __for__ loop to iterate over the dictionary keys and prints each key-value pair.




```python
# Define the function 'information' which takes a variable number of keyword arguments
def information(**kwargs):
    # Iterate over the keyword arguments and print each key-value pair
    for key, value in kwargs.items():  #
        print(key, ":", value)

# Call the 'information' function with 'name' and 'age' as keyword arguments
information(name = 'John Davis', age = 34)
```

```python
# Define the function 'information' which takes a variable number of keyword arguments
def information(**kwargs):
    # Iterate over the keyword arguments and print each key-value pair
    for key, value in kwargs.items():  # kwargs.keys(), kwargs.values(), kwargs.items()
        print(key, ":", value)

# Call the 'information' function with 'name' and 'age' as keyword arguments
information(name = 'John Davis', age = 34)
```

- This function call passes four keyword arguments: name, age, city, and country.
- The function prints the key-value pairs for all the arguments.

```python
# Call the 'information' function with multiple keyword arguments
information(name = 'John Davis', age = 34, city = 'London', country  = 'UK')
```

✅ Rule 1 — Positional-Only Parameters (/)
⭐ Meaning

Parameters before / can be passed ONLY by position, not by name.

Syntax
def f(a, b, /):

✅ Example
def demo(a, b, /):
    print(a, b)

demo(10, 20)        # ✅ correct
demo(a=10, b=20)    # ❌ error

⭐ Why useful

Used in:

built-in functions

performance APIs

when parameter name should not matter

Example:

len(obj)


You cannot write:

len(obj="abc") ❌

✅ Rule 2 — Positional-or-Keyword Parameters (Normal Parameters)

These are default Python parameters.

They can be passed:

by position

by keyword

Example
def demo(a, b):
    print(a, b)

demo(10, 20)        # positional
demo(b=20, a=10)    # keyword

✅ Rule 3 — *args (Variable Positional Arguments)
⭐ Meaning

Collects extra positional arguments into a tuple

Example
def demo(a, *args):
    print(a)
    print(args)

demo(10, 20, 30, 40)


Output:

a = 10
args = (20, 30, 40)

⭐ Important Insight

After *args, all parameters become keyword-only.

✅ Rule 4 — Keyword-Only Parameters

These parameters must be passed using name=value.

They appear:

after *args

OR after bare *

Example 1 (after *args)
def demo(*args, k1, k2):
    print(args, k1, k2)

demo(10, 20, k1=5, k2=6)   # ✅
demo(10, 20, 5, 6)         # ❌

Example 2 (using bare *)
def demo(a, *, k1, k2):
    print(a, k1, k2)

demo(10, k1=1, k2=2)   # ✅
demo(10, 1, 2)         # ❌

✅ Rule 5 — **kwargs (Variable Keyword Arguments)
⭐ Meaning

Collects extra keyword arguments into a dictionary

Example
def demo(a, **kwargs):
    print(a)
    print(kwargs)

demo(10, x=100, y=200)


Output:

a = 10
kwargs = {'x':100, 'y':200}

✅ MASTER ORDER RULE ⭐⭐⭐⭐⭐ (MOST IMPORTANT)

Function parameters must appear in this order:

1️⃣ Positional-only (/)
2️⃣ Normal parameters
3️⃣ *args
4️⃣ Keyword-only parameters
5️⃣ **kwargs

✅ FULL POWER EXAMPLE (All Rules Together)
def mega_func(a, b, /, c, d, *args, k1, k2, **kwargs):
    print("Positional-only:", a, b)
    print("Normal:", c, d)
    print("Args:", args)
    print("Keyword-only:", k1, k2)
    print("Kwargs:", kwargs)

Call
mega_func(
    1, 2,        # positional-only
    3, 4,        # normal
    5, 6, 7,     # *args
    k1=10, k2=20,
    x=100, y=200
)

🚨 Invalid Patterns (Very Important)
❌ **kwargs before *args
def f(**kwargs, *args):  ❌


Reason:
Python must first consume positional arguments.

❌ keyword-only before *args
def f(a, k1, *args):


Here k1 is NOT keyword-only.

To make keyword-only:

def f(a, *, k1):

✅ Real Framework Style Example (Very Useful)
def api_call(endpoint, /, timeout=10, *retries, headers=None, **options):


Used in:

HTTP clients

ML pipelines

decorators

orchestration systems

⭐ SUPER MEMORY HACK 🚀

Think like a funnel:

Positional-only
      ↓
Normal
      ↓
*args bucket
      ↓
Keyword-only gate
      ↓
**kwargs bucket


Arguments flow from left → right only.

⭐ FINAL INTERVIEW READY LINE

Python enforces a strict ordering of parameters: positional-only → positional-or-keyword → variable positional (*args) → keyword-only → variable keyword (**kwargs).



```python
def func(a, b, *args, **kwargs):
  print(a, b)
  print(args)
  print(kwargs)
func(1,2,3,4,5,x=7,y=8)
```

```python
def func(a, *args,c,d, **kwargs):
  print(a, d,c)
  print(args)
  print(kwargs)
func(1,2,3,c=4,d=5,x=7,y=8)
```
