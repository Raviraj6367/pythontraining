## __Generator functions__
A generator function allows iteration through values one at a time, without storing all the values in memory at once. This can be useful when dealing with large datasets or when only one value at a time is needed.

### __Define a generator function__
- This function uses yield to return values one at a time, pausing after each yield and resuming when next() is called again.
- Each time next() is called, the function continues from where it paused, and yields the next value.

```python
# Define the generator function 'myGenerator'
def myGenerator(n):
    # Initialize 'n' to 1
    n+=5 # Step 1
    # Print a message indicating the first iteration
    print('First Iteration')
    # Yield a string with the current value of 'n'
    yield 'Number 1 : n = ' + str(n) # soft return--> value is returned but the cursor stays at that line only.
# next()
    # Increment 'n' by 2
    n += 2 #Step 2
    # Print a message indicating the second iteration
    print('Second Iteration')
    # Yield a string with the current value of 'n'
    yield 'Number 2 : n = ' + str(n)

    # Increment 'n' by 3
    n += 100 # Step 3
    # Print a message indicating the third iteration
    print('Third Iteration')
    # Yield a string with the current value of 'n'
    yield 'Number 3 : n = ' + str(n)

    # Increment 'n' by 3
    n -=45 # Step 4.
    # Print a message indicating the third iteration
    print('Forth Iteration')
    # Yield a string with the current value of 'n'
    yield n
    n+=n
    return n
```

```python
g= myGenerator(100)
g
```

```python
g = myGenerator(100)

while True:
    try:
        print(next(g))   # prints all yield values
    except StopIteration as e:
        print("Return value =", e.value)   # prints return value
        break

        # We get the yielded values as well as return statement value
```

### __map function__
- The map() function applies the conversion function Fahrenhite to each temperature in the list temp.

```python
# Define a list of temperatures in Celsius
temp = [22.0, 45.3, 35.4, 40.0, 44.7, 21.5]
# Iterate over the list of temperatures
for T in temp:
    # Convert each temperature to Fahrenheit
    f = ((9/5) * T) + 32
    # Print the conversion result
    print('{:0.1f} in degree C is equivalent to {:0.1f} in degree F.'.format(T,f))
```

```python
# Define the function 'Fahrenhite' which converts Celsius to Fahrenheit
def Fahrenhite(T):
    # Return the converted temperature rounded to 1 decimal place
    return round(((9/5) * T) + 32, 1)
```

```python
results = (map(Fahrenhite, temp))# map--> Hey Fahrenhite FUnction,
# go ahead and apply yourself on each element of the temp list.
print('Temp in deg C ', temp)
print('Temp in deg F ', results)
```

### __Lambda function__
- A  lambda function is a small, anonymous function.
- It can be used inside map() without explicitly defining a separate function.

```python
def Farenhite(T):
  return round(((9/5) * T) + 32, 1)
```

```python
# Use the 'map' function with a lambda function to convert temperatures
results = list(map(lambda T: round(((9/5) * T) + 32, 1), temp))
# Lambda is an anonymous function.
# Print the original list of temperatures in Celsius
print('Temp in deg C ', temp)
# Print the converted list of temperatures in Fahrenheit
print('Temp in deg F ', results)
```

### __Filter function__
- The filter() function filters only odd numbers from the list by applying a condition (n % 2 != 0).

```python
# Define a list of numbers
numbers = [32, 45, 62, 21, 55, 91, 88, 17 ]
# Initialize an empty list to store odd numbers
odd_numbers = []
# Iterate over the list of numbers
for n in numbers:
    # Check if the number is odd
    if n % 2 !=0 :
        # Append the odd number to the list
        odd_numbers.append(n)
# Print the list of odd numbers
print(odd_numbers)
```

```python
# Define a list of numbers
numbers = [32, 45, 62, 21, 55, 91, 88, 17 ]
# Use the 'filter' function with a lambda function to filter odd numbers
odd_numbers = list(filter ( lambda n : n%2!=0,  numbers)) # return type of filter is bool

# Print the list of odd numbers
print(odd_numbers)
```

### __Reduce function__
- The reduce() function implements the logic and reduces the elements in a data structure to a single value.

```python
# Define a list of numbers
numbers = [2,4,6,8,10,12]
# Initialize a variable 'tot' to store the total
tot = 0
# Iterate over the list of numbers
for n in numbers:
    # Add each number to 'tot'
    tot += n
# Print the total sum
print('Total :', tot)
```

```python
# Import the 'reduce' function from 'functools'
from _functools import reduce
# Define a list of numbers
numbers = [2,4,6,8,10,12]
# Use the 'reduce' function with a lambda function to calculate the total sum
result = reduce(lambda x , y : x + y, numbers)
# Print the total sum
print('Total = ', result)

# Itr 1 : take 1st 2. numbers from the list.  (2,4) --> 2+4 = 6 ,
#This 6 is stored in x.
#Itr 2: next no. is the list and assign it to y --> (6,6)= 6+6=12,
# 12 is soored in x.
# Itr 3....
```

```python
def interactive_counter():
    count = 0
    while True:
        # yield returns 'count', and (yield count) evaluates to the value sent in
        increment_amount = yield count
        print(str(increment_amount), str(count))
        if increment_amount is not None:
            count += increment_amount
        else:
            count += 1

#yield-->0
# result of the expression will hold the value sent by send. inclrement_amount=510.
#COUNT= 510
# YEILDS= 510
```

```python
gen = interactive_counter()
print(f"Initial count: {next(gen)}")
print(f"Initial count: {next(gen)}")
print(f"Initial count: {next(gen)}")
print(f"Count after sending +5: {gen.send(510)}")
# send(value) sends a value into a generator at the point where it last yielded.
print(f"Initial count: {next(gen)}")
print(f"Initial count: {next(gen)}")
print(f"Count after sending +5: {gen.send(12)}")
print(f"Initial count: {next(gen)}")
```

```python
def running_total():
    total = 0
    while True:
        x = yield total
        print('Printing x and total' ,str(x), str(total))
        total += x
```

```python
g = running_total()

print(next(g))      # start generator → 0
print(g.send(10))   # 10
print(g.send(5))    # 15
print(g.send(20))   # 35
```
